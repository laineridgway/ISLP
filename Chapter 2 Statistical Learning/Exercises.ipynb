# c)

```python
def print_mean_std(name, series):
    mean_value = np.mean(series)
    std_value = np.std(series)
    print(f"{name:<12}: Mean: {mean_value:7.2f}, Std: {std_value:.2f}")
```

```python
quant_series = auto.select_dtypes(include=['number']).columns.tolist()
```

```python
for i in quant_series:
    print_mean_std(i, auto[i])
```

```python
# Drop the 10th through 85th observations
auto_subset = auto.drop(x for x in range(10, 85) if x != 32)

# Compute the range, mean, and standard deviation for each column
summary = auto_subset.describe()
summary.loc['range'] = summary.loc['max'] - summary.loc['min']

# Print the summary
print(summary)
```
