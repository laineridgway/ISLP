```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
college = pd.read_csv('../ISLP_Data/College.csv')
```

```python
college2 = pd.read_csv('../ISLP_Data/College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'},
axis=1)
college3 = college3.set_index('College')
```

```python
college = college3
```

```python
C = ['Top10perc', 'Apps', 'Enroll']
pd.plotting.scatter_matrix(college[C])
```

```python
college.boxplot(column='Outstate', by='Private')
```

```python
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])
```

```python
college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
college['Outstate'].plot.hist(ax=axs[0, 0], bins=30)
college['Accept'].plot.hist(ax=axs[0, 1], bins=30)
college['Enroll'].plot.hist(ax=axs[1, 0], bins=30)
college['Expend'].plot.hist(ax=axs[1, 1], bins=30)
```

```python
# Select a subset of variables
vars_of_interest = ['Apps', 'Accept', 'Enroll', 'Top10perc', 'Expend']

sns.pairplot(college[vars_of_interest])
plt.show()
```
